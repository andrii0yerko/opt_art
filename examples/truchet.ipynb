## Transform

```python
from opt_art import ElasticTruchetTiles

tiler = ElasticTruchetTiles()
```

Transform method takes an image and two parameters:

`block_size` - window length to be replaced with a single tile. Keep it reasonable small

`pattern` - generator pattern for a Truchet tiles. Should be either "random", string of space separated equal-sized words or a square matrix consists of the following letters:

- **a** - bottom left - 
- **b** - top left - 
- **c** - top right - 
- **d** - bottom left - 

So pattern **"da dc"** will looks like this:


```python
tiler.transform(img, 5, "ac ca")
```

```python
tiler.transform(img, 5, "ca db")
```

```python
tiler.transform(img, 5, "da cb")
```

```python
tiler.transform(img, 5, "bddbddaacaac dbddbdacaaca ddbddbcaacaa bddbddaacaac dbddbdacaaca ddbddbcaacaa ccaccadbbdbb caccacbdbbdb accaccbbdbbd ccaccadbbdbb caccacbdbbdb accaccbbdbbd")
```

```python
tiler.transform(img, 5, "random")
```
