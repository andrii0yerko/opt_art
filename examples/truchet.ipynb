## Transform

```python
from opt_art import ElasticTruchetTiles

tiler = ElasticTruchetTiles()
```

The transform method takes an image and two parameters:

`block_size` - window length to be replaced with a single tile. Keep it reasonably small

`pattern` - generator pattern for Truchet tiles. Should be either "random", a string of space-separated equal-sized words or a square matrix consisting of the following letters:

- **a** - bottom left - 
- **b** - top left - 
- **c** - top right - 
- **d** - bottom left - 

So pattern **"dc ab"** will look like this:


```python
tiler.transform(img, 4, "ac ca")
```

```python
tiler.transform(img, 3, "ca db")
```

```python
tiler.transform(img, 3, "da cb")
```

```python
tiler.transform(img, 3, "bddbddaacaac dbddbdacaaca ddbddbcaacaa bddbddaacaac dbddbdacaaca ddbddbcaacaa ccaccadbbdbb caccacbdbbdb accaccbbdbbd ccaccadbbdbb caccacbdbbdb accaccbbdbbd")
```

```python
tiler.transform(img, 4, "random")
```
