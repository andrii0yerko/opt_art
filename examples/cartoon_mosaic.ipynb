# Cartoon Mosaic Tiles

Tiling an image with equal number of each unique element

## Load an Image

The ideal input for this algorithm - a contrast image, which has almost equal-squared areas of different brightnesses

```python
import glob
from PIL import Image

img = Image.open("./data/image-2.jpg")
img
```

## Transform

`CartoonMosaics` constructor takes an iterable of images used as unique tiles in the resulting mosaic. Each of them will appear the same number of times

For tiles choose 3-7 images with different brightness levels. In this example, we will use clips from the "Mr.Incredible becoming uncanny" meme - you may use any other set or just shades of gray.

```python
from opt_art import CartoonMosaics

filenames = glob.glob("./data/mrincredible/*.jpg")
mrincredible_images = [Image.open(fn) for fn in filenames]

tiler = CartoonMosaics(mrincredible_images)
```

`CartoonMosaics.transform` method takes an image and an `approx_block_size` parameter: the lower it is, the more detailed image you will get, but it takes more time to find solution.

```python
tiler.transform(img, 15)
```
